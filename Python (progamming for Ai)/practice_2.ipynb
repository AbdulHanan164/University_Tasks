```python
class TextAnalyzer:
    def __init__(self, filename):
        self.filename = filename
        self.word_count = {}
    
    def analyze(self):
        """Analyze the text file and count the occurrences of each word."""
        with open(self.filename, 'r') as file:
            for line in file:
                words = line.split()
                for word in words:
                    word = word.lower()  # Convert to lowercase to ensure case-insensitive counting
                    self.word_count[word] = self.word_count.get(word, 0) + 1
        
    def top_n_words(self, n=10):
        """Get the top N most frequent words."""
        sorted_words = sorted(self.word_count.items(), key=lambda x: x[1], reverse=True)
        return sorted_words[:n]
    
    def display_top_n_words(self, n=10):
        """Display the top N most frequent words."""
        top_words = self.top_n_words(n)
        print(f"Top {n} most frequent words:")
        for word, count in top_words:
            print(f"{word}: {count} times")


def main():
    # Create a TextAnalyzer instance with a specific filename
    path=open("file.txt","w")
    string=str(input("Enter string :"))
    path.write(string)
    analyzer = TextAnalyzer("file.txt")  # Assume you have a file named "sample_text.txt"

    # Analyze the text file
    analyzer.analyze()

    # Display the top 10 most frequent words
    analyzer.display_top_n_words(10)


if __name__ == "__main__":
    main()
```

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def __add__(self, other):
        """Overload the addition operator (+)"""
        return Vector(self.x + other.x, self.y + other.y)
    
    def __sub__(self, other):
        """Overload the subtraction operator (-)"""
        return Vector(self.x - other.x, self.y - other.y)
    
    def __mul__(self, other):
        """Overload the multiplication operator (*)"""
        return Vector(self.x * other.x, self.y * other.y)
    
    def __str__(self):
        """String representation of the Vector object"""
        return f"({self.x}, {self.y})"
    
    def __eq__(self, other):
        """Overload the equality operator (==)"""
        return self.x == other.x and self.y == other.y

# Create Vector objects
v1 = Vector(5, 3)
v2 = Vector(8, 5)

# Add, subtract, and multiply Vectors and print the results
v3 = v1 + v2
print("Addition:", v3)

v4 = v1 - v2
print("Subtraction:", v4)

v5 = v1 * v2
print("Multiplication:", v5)

# Check equality
print("Equality:", v1 == Vector(6, 7))
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris=load_iris()
x=iris.data
y=iris.target
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4,random_state=1,stratify=y)
from sklearn.preprocessing import StandardScaler
scalar=StandardScaler()
x_train=pd.DataFrame(scalar.fit_transform(x_train),columns=iris.feature_names)
x_test=pd.DataFrame(scalar.transform(x_test),columns=iris.feature_names)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
from sklearn.model_selection import cross_val_score
scores=cross_val_score(model,x_train,y_train,cv=10)
print(np.mean(scores))
```

```python
df_coef=pd.DataFrame(model.coef_,columns=x_train.columns)
df_coef
```

```python
predictions = model.predict(x_test)

# Create a DataFrame to compare actual vs predicted values
compare_df = pd.DataFrame({'actual': y_test, 'predicted': predictions})

compare_df
```
