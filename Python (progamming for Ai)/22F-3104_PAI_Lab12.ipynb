```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0, 10)
y = x**2
plt.plot(x, y, color='b', linestyle='--')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Plot')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.random.rand(20)
y = np.random.rand(20)
sizes = np.random.randint(10, 100, 20)
sizes_normalized = sizes / sizes.max()
plt.scatter(x, y, s=sizes_normalized * 100, cmap='viridis', alpha=0.7)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.colorbar(label='Sizes')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
categories = ['Aplha', 'Beta', 'Gemma']
values = np.array([12, 34, 43])
errors = np.array([21, 25, 33])
plt.bar(categories, values, yerr=errors, color='khaki', capsize=7, edgecolor='black', lw=0.8)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Error Bars')
plt.grid(axis='y', linestyle='--', alpha=1)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Error Bars')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.random.randn(1000)
plt.hist(data, bins=20, color='orange', edgecolor='black')
plt.hist(data, bins=30, color='khaki', alpha=0.7, edgecolor='black')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Overlaped Histograms chart')
plt.legend(['Orange Histogram', 'Khaki Histogram'])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-10, 10, 100)
y1 = x
y2 = x**2
y3 = x**3
y4 = np.sin(x)
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y1, label='y = x', color='blue')
axes[0, 0].set(title='Linear Function', xlabel='X-axis', ylabel='Y-axis')
axes[0, 0].legend()
axes[0, 1].plot(x, y2, label='y = x^2', color='green')
axes[0, 1].set(title='Quadratic Function', xlabel='X-axis', ylabel='Y-axis')
axes[0, 1].legend()
axes[1, 0].plot(x, y3, label='y = x^3', color='red')
axes[1, 0].set(title='Cubic Function', xlabel='X-axis', ylabel='Y-axis')
axes[1, 0].legend()
axes[1, 1].plot(x, y4, label='y = sin(x)', color='purple')
axes[1, 1].set(title='Sine Function', xlabel='X-axis', ylabel='Y-axis')
axes[1, 1].legend()
fig.suptitle('Different Mathematical Functions')
plt.tight_layout()
plt.show()
```
