```python
str_input = input("Input the string: ")
# Create an empty dictionary to count characters
char_count = {}
for char in str_input:
    # Check if the character is a letter lowercase or uppercase
    if ('a' <= char <= 'z' or 'A' <= char <= 'Z' ):       
        # If the character is not already in the dictionary, initialize its count to 0
        char_count[char] = char_count.get(char, 0) + 1

# Initialize two variables to count even and odd characters
even_count = 0
odd_count = 0

for count in char_count.values():
    # Check if the frequency is even, else it's odd
    if count % 2 == 0:       
        even_count += 1
    else:
        odd_count += 1
# Check if the number of characters with even frequency is equal to the total number of characters in the string
if even_count == len(char_count):    
    print("0")
elif odd_count ==len(char_count):
    # If at least one character has odd frequency the string is odd.
    print("1")
else:
    # Print string is neither even nor odd.
    print("2")
```

```python
credit_card = input("Enter Credit card number: ")
num_list = []
double_list = []
total = 0
# Iterate over the credit card number and append each digit to num_list
for i in range(0, len(credit_card)):
    num_list.append(credit_card[i])
# Slice from the right side by using negative indexing
slice_list1 = num_list[-2::-2]
slice_list2 = num_list[-1::-2]
# Double the digits in slice_list1 and add them to double_list
for i in slice_list1:
    double_num = int(i) * 2
    if double_num >= 10:
        double_num -= 9
    double_list.append(str(double_num))  # Convert the result back to a string
# Calculate the sum of the digits in double_list and the digits in slice_list2
total = sum(int(digit) for digit in double_list) + sum(int(digit) for digit in slice_list2)
print("Total:", total)
# Check if the total is a multiple of 10
# Print the entire credit card number without commas and spaces
print(''.join(num_list),end=' ')
if total % 10 == 0:
    print("Credit card number is valid.")
else:
    print("Credit card number is invalid.")
```

```python
n = int(input("Enter the number of values to check: "))

for _ in range(n):
    value = int(input())
    print(f"Input value: {value}")

    count = 0
    for i in range(1, value + 1):
        if value % i == 0:
            count += 1

    if count == 2:
        print(f"Would you believe it; {value} is a prime!")
    else:
        # Find the closest smaller prime
        smaller_prime = value - 1
        count = 0
        while smaller_prime >= 2:
            if value % smaller_prime == 0:
                count += 1
            if count == 2:
                break
            smaller_prime -= 1

        # Find the closest larger prime
        larger_prime = value + 1
        count = 0
        while True:
            for i in range(2, int(larger_prime ** 0.5) + 1):
                if larger_prime % i == 0:
                    count += 1
                    break
            if count == 0:
                break
            count = 0
            larger_prime += 1

        # Calculate how close the number is to a prime
        if value - smaller_prime <= larger_prime - value:
            print(f"Missed it by that much ({value - smaller_prime})! The closest prime is {smaller_prime}.")
        else:
            print(f"Missed it by that much ({larger_prime - value})! The closest prime is {larger_prime}.")

    print() 
```

```python
# Part A: Flattening a Nested List
input_list = [1, 2, 3, [54, 5, 6, 8], [56, 44], [1, 1, 1, 1, 1, 1], 8, 9, 10]

# Flatten the nested list using list comprehension
flattened_list = [num for sublist in input_list for num in (sublist if isinstance(sublist, list) else [sublist])]

# Print the flattened list
print(flattened_list)

# Part B: Word Frequency Analysis
text = "Programmming is fun."

# Split the text into words (tokenization) and remove common punctuation characters
punctuation = ".,/!?@[]{}():';\""
words = text.lower().split()
word_freq = {}

for word in words:
    # Remove punctuation characters from the beginning and end of each word
    while word and word[0] in punctuation:
        word = word[1:]
    while word and word[-1] in punctuation:
        word = word[:-1]
    
    # Update word frequencies in the dictionary
    if word:
        if word in word_freq:
            word_freq[word] += 1
        else:
            word_freq[word] = 1

# Sort the word-frequency pairs by frequency in descending order
sorted_words = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Print the result
for word, freq in sorted_words:
    print(f"{word}: {freq}")

# Part C: Matrix Multiplication
def matrix_multiply(matrix1, matrix2):
    # Check if matrix multiplication is valid
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Matrix dimensions are not compatible for multiplication")
    # Perform matrix multiplication using list comprehension
    result = [[sum(matrix1[i][k] * matrix2[k][j] for k in range(len(matrix2))) for j in range(len(matrix2[0]))] for i in range(len(matrix1))]
    return result
# Example usage:
matrix1 = [[1, 2, 3], [4, 5, 6]]
matrix2 = [[7, 8], [9, 10], [11, 12]]

result = matrix_multiply(matrix1, matrix2)

# Print the result
for row in result:
    print(row)
```
