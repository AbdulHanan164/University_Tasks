```python
import pandas as pd
import matplotlib.pyplot as plt
# Part a)Read and Extract sales data for face cream and face wash
data = pd.read_csv("company_sales_data.csv")
face_cream_data = data[['month_number', 'facecream']]
face_wash_data = data[['month_number', 'facewash']]

# Part b) Create a bar chart for face cream
plt.figure(figsize=(10, 6))
plt.bar(face_cream_data['month_number'], face_cream_data['facecream'], color='blue', label='Face Cream', alpha=0.7)
plt.bar(face_wash_data['month_number'], face_wash_data['facewash'], color='green', label='Face Wash', alpha=0.7)
# Add labels to each bar
for i, value in enumerate(face_cream_data['facecream']):
    plt.text(face_cream_data['month_number'][i], value + 50, str(value), ha='center', va='bottom')
for i, value in enumerate(face_wash_data['facewash']):
    plt.text(face_wash_data['month_number'][i], value + 50, str(value), ha='center', va='bottom')
# Customize the chart
plt.title('Sales Performance of Face Cream and Face Wash Over the Last Year')
plt.xlabel('Month')
plt.ylabel('Sales Units')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load employees.csv
employees_data = pd.read_csv("employees.csv")

# Task 1
def display_overview():
    print(employees_data.head())
    print(employees_data.describe())
# Task 2

def handle_missing_duplicates():
    print("Missing values before handling:")
    print(employees_data.isnull().sum())

    # Replace missing values based on data type
    for column in employees_data.columns:
        if employees_data[column].dtype in ['int64', 'float64']:
            # Replace missing values in numerical columns with the mean
            employees_data[column].fillna(employees_data[column].mean(), inplace=True)
        else:
            # Replace missing values in categorical columns with the mode
            employees_data[column].fillna(employees_data[column].mode().iloc[0], inplace=True)

    # Remove duplicates
    employees_data.drop_duplicates(inplace=True)

    print("Missing values after handling:")
    print(employees_data.isnull().sum())
# Task 3
def calculate_average_highest_salary():
    print("Average Salary:", employees_data['SALARY'].mean())
    print("Employee with the Highest Salary:")
    print(employees_data.loc[employees_data['SALARY'].idxmax()])

# Task 4
def employees_per_department():
    print("Number of Employees in Each Department:")
    print(employees_data['DEPARTMENT_ID'].value_counts())
    print("\nAverage Salary for Each Department:")
    print(employees_data.groupby('DEPARTMENT_ID')['SALARY'].mean())

# Task 5
def convert_hire_date_identify_recent_hires():
    employees_data['HIRE_DATE'] = pd.to_datetime(employees_data['HIRE_DATE'])
    print("Employees Hired in the Last Year:")
    print(employees_data.loc[employees_data['HIRE_DATE'] > pd.to_datetime('today') - pd.DateOffset(years=1)])

# Task 6
def filter_sort_IT_department_visualize_data():
    it_department_data = employees_data[employees_data['DEPARTMENT_ID'] == 60]
    sorted_it_data = it_department_data.sort_values(by='SALARY', ascending=False)
    print("Employees in IT Department, sorted by Salary:")
    print(sorted_it_data)

    # Bar chart showing the number of employees in each department
    employees_data['DEPARTMENT_ID'].value_counts().plot(kind='bar', title='Number of Employees in Each Department')
    plt.show()

    # Histogram to visualize salary distribution
    employees_data['SALARY'].plot(kind='hist', bins=20, title='Salary Distribution')
    plt.show()

# Task 7
def total_salary_expenditure_identify_highest_expenditure_department():
    total_salary_expenditure = employees_data.groupby('DEPARTMENT_ID')['SALARY'].sum()
    print("Total Salary Expenditure for Each Department:")
    print(total_salary_expenditure)
    highest_expenditure_department = total_salary_expenditure.idxmax()
    print("Department with the Highest Total Salary Expenditure:", highest_expenditure_department)

# Menu-driven script
while True:
    print("\n=== Menu ===")
    print("1. Display Overview")
    print("2. Handle Missing Values and Remove Duplicates")
    print("3. Calculate Average Salary and Identify Highest Salary")
    print("4. Employees per Department and Average Salary")
    print("5. Convert 'HireDate' and Identify Recent Hires")
    print("6. Filter and Sort IT Department Visualize Data")
    print("7. Total Salary Expenditure and Identify Highest Expenditure Department")
    print("8. Exit")

    choice = input("Enter your choice (1-8): ")

    if choice == '1':
        display_overview()
    elif choice == '2':
        handle_missing_duplicates()
    elif choice == '3':
        calculate_average_highest_salary()
    elif choice == '4':
        employees_per_department()
    elif choice == '5':
        convert_hire_date_identify_recent_hires()
    elif choice == '6':
        filter_sort_IT_department_visualize_data()
    elif choice == '7':
        total_salary_expenditure_identify_highest_expenditure_department()
    elif choice == '8':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 8.")
```

```python
import pandas as pd
import folium
from folium.plugins import MarkerCluster
import seaborn as sns
import matplotlib.pyplot as plt

def read_csv_and_create_map():
    """
    Task 1 and 2: Read CSV and create a map with office locations.
    """
    df = pd.read_csv("fatalities.csv", parse_dates=["date_of_event", "date_of_death"])
    df.columns = df.columns.str.strip()

    map_center = [df['latitude'].mean(), df['longitude'].mean()]
    mymap = folium.Map(location=map_center, zoom_start=10)

    for index, row in df.iterrows():
        popup_text = f"{row['name']} - {row['place_of_residence']}, {row['place_of_residence_district']}\n" \
                     f"Date of Event: {row['date_of_event'].strftime('%d/%m/%Y')}\n" \
                     f"Date of Death: {row['date_of_death'].strftime('%d/%m/%Y')}\n" \
                     f"Age: {row['age']} | Gender: {row['gender']}\n" \
                     f"Notes: {row['notes']}"
        folium.Marker(location=[row['latitude'], row['longitude']], popup=popup_text).add_to(mymap)

    marker_cluster = MarkerCluster().add_to(mymap)

    for index, row in df.iterrows():
        popup_text = f"{row['name']} - {row['place_of_residence']}, {row['place_of_residence_district']}\n" \
                     f"Date of Event: {row['date_of_event'].strftime('%d/%m/%Y')}\n" \
                     f"Date of Death: {row['date_of_death'].strftime('%d/%m/%Y')}\n" \
                     f"Age: {row['age']} | Gender: {row['gender']}\n" \
                     f"Notes: {row['notes']}"
        folium.Marker(location=[row['latitude'], row['longitude']], popup=popup_text).add_to(marker_cluster)

    mymap.save("fatalities_map.html")
    print("Fatalities map created and saved.")

def visualize_injuries():
    """
    Task b: Visualize types of injuries using seaborn.
    """
    df = pd.read_csv("fatalities.csv", parse_dates=["date_of_event", "date_of_death"])
    df.columns = df.columns.str.strip()

    injury_types = df['type_of_injury'].value_counts()

    plt.figure(figsize=(10, 6))
    sns.barplot(x=injury_types.index, y=injury_types.values, palette="viridis")
    plt.title("Types of Injuries")
    plt.xlabel("Injury Type")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.show()

def analyze_means_used():
    """
    Task c: Analyze means used in fatalities using seaborn.
    """
    df = pd.read_csv("fatalities.csv", parse_dates=["date_of_event", "date_of_death"])
    df.columns = df.columns.str.strip()

    means_used_counts = df['ammunition'].value_counts()

    plt.figure(figsize=(12, 6))
    sns.barplot(x=means_used_counts.index, y=means_used_counts.values, palette="plasma")
    plt.title("Means Used in Fatalities")
    plt.xlabel("Means Used")
    plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.show()

while True:
    print("\nMenu:")
    print("1. Read CSV and create fatalities map.")
    print("2. Visualize types of injuries.")
    print("3. Analyze means used in fatalities.")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    if choice == '1':
        read_csv_and_create_map()
    elif choice == '2':
        visualize_injuries()
    elif choice == '3':
        analyze_means_used()
    elif choice == '4':
        print("Exit the program(^_^)| Goodbye!")
        break
    else:
        print("Invalid choice. Please enter a number between 1 and 4.")
```

```python
import requests
from bs4 import BeautifulSoup
def fetch_html_content(url):
    response = requests.get(url)
    return response.text
def parse_html(html_content):
    return BeautifulSoup(html_content, "html.parser")
def extract_data(soup):
    headlines = []
    links = []
    # Find all <a> elements within the <h2> elements
    headline_elements = soup.select("h2 a")
    for headline in headline_elements:
        headlines.append(headline.text.strip())
        links.append(headline.get("href"))
    return headlines, links
def display_information(data, label):
    print(f"\n{label.capitalize()}:")
    for i, item in enumerate(data, start=1):
        print(f"{i}. {item}")
# URL of the news website
url = "https://www.dawn.com/"
while True:
    print("\n=== Menu ===")
    print("1. Fetch and Display News Headlines")
    print("2. Fetch and Display News Links")
    print("3. Exit")
    choice = input("Enter your choice (1, 2, or 3): ")
    if choice == '1':
        html_content = fetch_html_content(url)
        soup = parse_html(html_content)
        headlines, _ = extract_data(soup)
        display_information(headlines, "headlines")
    elif choice == '2':
        html_content = fetch_html_content(url)
        soup = parse_html(html_content)
        _, links = extract_data(soup)
        display_information(links, "links")
    elif choice == '3':
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please enter 1, 2, or 3.")
```
