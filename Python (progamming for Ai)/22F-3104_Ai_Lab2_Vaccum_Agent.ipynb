```python
import random

class VacuumCleaner:
    def __init__(self):
        self.room = [['left'], ['right']]  # for directions of room
        self.cleaning_performance = 0  # for checking its detection ability

    def clean_room(self):  # for checking the number of cleaned boxes
        cleaned_count = 0
        start_side = input("Enter the side to start from ('left' or 'right'): ")
        for rooms in range(0, self.num_rooms):  # loop for getting data, randomly on where dirt is located
            if rooms > 0:
                start_side = random.choice(['left', 'right'])  # randomly select side after the first room

            while start_side not in ['left', 'right']:  # loop for naming the direction wrong the first time
                print("Invalid input. Please enter 'left' or 'right'.")
                start_side = input("Enter the side to start from ('left' or 'right'): ")

            dirt_locations = self.sensors_for_detection()
            cleaned_count += self.cleaning_with_actuators(dirt_locations, rooms)

        self.cleaning_performance = (cleaned_count / (2 * self.num_rooms)) * 100  # performance calculation
        # status of our Vacuum Cleaner
        print('Status: All rooms are now clean.')
        print('Room: ', self.room)
        print('Cleaning needed to perform: ', self.cleaning_performance, '%')

    def sensors_for_detection(self):  # this part of code will do the task of sensors, where it will detect the dirt
        return [random.choice(['clean', 'dirty']) for i in range(2)]

    def cleaning_with_actuators(self, dirt_locations,
                                room_number):  # now this part of code will perform work on the dirt
        cleaned_count = 0
        for i in range(2):  # for telling status in which room vacuum cleaner is working
            print(f"Working in room {room_number + 1}: {self.room[i][0]}'s side")
            if dirt_locations[i] == 'clean':
                print(f"Room number {room_number + 1} is clean. Moving to the other side.")
                print()
            else:
                print(f"Room number {room_number + 1} is dirty. Sucking dirt.")
                dirt_locations[i] = 'clean'
                cleaned_count += 1
                print()
        return cleaned_count

    def set_num_rooms(self, num_rooms):
        self.num_rooms = num_rooms


if __name__ == "__main__":
    cleaner = VacuumCleaner()
    num_rooms = int(input("Enter the number of rooms to clean: "))
    cleaner.set_num_rooms(num_rooms)
    cleaner.clean_room()
```
