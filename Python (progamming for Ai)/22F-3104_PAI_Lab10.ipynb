```python
#Question 3 Part(a)
import matplotlib.pyplot as plt
import pandas as pd
class_survival_rate = df.groupby('Pclass')['Survived'].mean()
plt.figure(figsize=(10, 6))
plt.bar(class_survival_rate.index, class_survival_rate.values, color='skyblue',width=0.2)
plt.title('Survival Rate Based on Ticket Class')
plt.xlabel('Ticket Class')
plt.ylabel('Survival Rate')
plt.show()
```
